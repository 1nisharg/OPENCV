## BASIC IMAGE MANIPULATION

```python
#Accessing Individual Pixels
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import Image
%matplotlib inline


ns_img = cv2.imread("VK18.png",0)

#printing the pixels
print(ns_img[0,0])
print(ns_img[0,6])
```

#### Modifiying img pixels


```python
ns_img_copy = ns_img.copy()
ns_img_copy[2,2] = 200
ns_img_copy[2,3] = 200
ns_img_copy[3,2] = 200
ns_img_copy[3,3] = 200

#alternate method --> ns_img_copy[2:3,2:3] = 200

plt.imshow(ns_img_copy, cmap="gray")
print(ns_img_copy)
```

#### Cropping Images

```python
img_crop = cv2.imread("NS.jpg", cv2.IMREAD_COLOR)
img_crop = img_crop[:,:,::-1]
plt.imshow(img_crop)
```

```python
cropped_region = img_crop[0:150, 150:350]
plt.imshow(cropped_region)
```

#### Resizing Images


Function Syntax:-

dst = resize(src, dsize[, dst[, fx[, fy[, interpolation] ] ] ] )


It has **2 required arguments**

1. src: input image
2. dsize: output image size

**Optional Arguments**
1. fx: scale actor along the horizontal axis
2. fy: scale factor along the vertical axis


```python
resized_cropped = cv2.resize(cropped_region,None,fx=2,fy=2)
plt.imshow(resized_cropped)
```

```python
#Method-2 Specifying exact size of the output image

desired_width = 100
desired_height = 200

dim=(desired_width, desired_height)

#resize bg img to to see size as logo image
resized_cropped = cv2.resize(cropped_region, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized_cropped)
```

```python
#Method-3 Resize while maintaining aspect ratio
desired_width=100
aspect_ratio = desired_width/cropped_region.shape[1]
desired_height = int(cropped_region.shape[0]*aspect_ratio)
dim = (desired_width, desired_height)

#Resize Image
resized_cropped = cv2.resize(cropped_region, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized_cropped)
```

```python
#FLIPPING THE IMAGE

img_flip = cv2.flip(cropped_region, 1)
plt.imshow(img_flip)
```
